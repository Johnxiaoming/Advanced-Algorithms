```
def CoinsChangeMoney(coins, money):
    result = []
    for i in coins:
        if money % i == 0:
            result.append(money // i)
    coins_seq = {}
    while money != 0:
        coins_seq[max(coins)] = money // max(coins)
        money = money % max(coins)
        coins.remove(max(coins))
    result.append(sum(coins_seq.values()))
    return min(result)
```
