```
def read_data(file_name):
    with open(file_name, 'r+') as file:
        line = file.readline().strip()

    line = line[1:len(line)-1].split()
    data = []
    for item in line:
        data.append(item)

    return data
```

```
def rev_subseq(start, end, str):
    seq = str[:start]
    seq += str[start:end][::-1]
    seq += str[end:]
    for i in range(start, end):
        if "+" in seq[i]:
            seq[i] = seq[i].replace("+", "-")
        else:
            seq[i] = seq[i].replace("-", "+")

    return seq
```

```
def format_answ(data):
    answ = " ".join(data)
    answ = "(" + answ + ")"
    return answ
```

```
def greedy_sort_rev(data):
    answer = []
    for i in range(1, len(data)+1):
        for j in range(i-1, len(data)):
            if str(i) == data[j][1:]:
                data = rev_subseq(i-1, j+1, data)
                answer.append(format_answ(data))
                if "-" in data[i-1]:
                    data[i-1] = data[i-1].replace("-", "+")
                    answer.append(format_answ(data))

                break

    return answer
```

```
def write_to_file(file_name, data):
    with open(file_name, 'w') as file:
        for i in range(len(data)):
            file.writelines(data[i]+"\n")

if __name__ == "__main__":
    data = read_data("/content/drive/My Drive/Colab Notebooks/Rosalind/rosalind_ba6a.txt")
    answ = greedy_sort_rev(data)
    write_to_file("/content/drive/My Drive/Colab Notebooks/ba6a_result.txt", answ)
```

```
###Another way

f = open('/content/drive/My Drive/Colab Notebooks/Rosalind/rosalind_ba6a.txt', 'r')
sample = f.readline().rstrip()
sample = sample.replace('(','').replace(')','')
sample = sample.split()
sample = [int(i) for i in sample]

def change_sign(sample):
    """
    Changes sign of number from:(+ -> -) and (- -> +)
    """
    for i in range(len(sample)):
        sample[i] = sample[i] * -1
        
def find_index(sample, next_num):
    """
    Obtains the index of the next number in the greedy algorithm.
    Determined based on next_num + 1, since our range begins at 0 and
    algorithm begins at 1.
    """
    index = 0
    if (next_num+1) in sample:
        index = sample.index(next_num+1)
    elif -(next_num+1) in sample:
        index = sample.index(-(next_num+1))
    return index

def output_format(sample):
    """
    Prints out the desired output format:
        (+1 -3 +4 -2 -5)
    """
    p = sample[::]
    for i in range(len(p)):
        p[i] = '{0:+g}'.format(p[i])
    p = str(p)
    p = p.replace('[', '(').replace(']',')').replace(',','').replace("'", '')
    print(p)

                
# Greedy Algorithm
def main():
    for i in range(len(sample)):
        if abs(sample[i]) != i+1:
            index = find_index(sample, i)
            sub = sample[i:index+1]
            sub = sub[::-1]
            change_sign(sub)
            sample[i:index+1] = sub
            output_format(sample)
        if sample[i] < 0:
            sample[i] = sample[i] * -1
            output_format(sample)
        
        
if __name__ == '__main__':
    main()
```
